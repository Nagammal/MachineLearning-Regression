```python
#Process Name : Data Collection & Data Preprocessing.pandas is a python library which is used for data handling and transformation
import pandas as pd 
#print(dir(pd)) 
```

```python
#read_csv is a fucntion inside pandas which will read/load the comma separated data.
data=pd.read_csv("50_Startups.csv") 
data.columns
```

```python
data=pd.get_dummies(data,drop_first=True,dtype=int)
data
```

```python
#Process Name : Input Output Split
independent=data[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida','State_New York']] #Assigning the dataset column to independent which is input
dependent=data[['Profit']]#Assigning the dataset column to dependent which is output/target
```

```python
#Process Name : Train Test Split
from sklearn.model_selection import train_test_split #sklearn is a library where all ML algorithms are present.Here we are importing function called train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.3, random_state=0) #Splitting the dataset in to test and train.test set being 30%(0.3)
```

```python
#Process Name : Model Creation
#Importing/Creating Decision tree model
from sklearn.tree import DecisionTreeRegressor
Model=DecisionTreeRegressor(criterion='absolute_error', max_features='log2',splitter='random', min_samples_split=10,max_depth=10)
Model.fit(X_train,Y_train) #giving the training set to the model
```

```python
#Import library to graphically view the tree model created
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(Model)
#plt.show()
```

```python
#Process Name : Prediction
Y_predicted=Model.predict(X_test)
Y_predicted
```

```python
#Process Name : Evaluation metrics
from sklearn.metrics import r2_score  #r2_score fucntion in sklearn is used to evaluate model
r_score=r2_score(Y_test,Y_predicted)
print("r_score/accuracy is :",r_score)
```

```python
#Process Name : Saving the model
import pickle 
filename="finalized_decision_tree_regression_model.sav"  #model is saved in the filename
pickle.dump(Model,open(filename,'wb')) #assigning the model to this file given as write binary
```
